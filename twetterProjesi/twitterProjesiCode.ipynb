```python
location_dict = {}
for loc in new_df["user_location"].unique():
    if "," in loc:
        country = loc.split(",")[-1].strip()
    else:
        country = loc.strip()
    if country in location_dict:
        location_dict[country].append(loc)
    else:
        location_dict[country] = [loc]
```

```python
!pip install fuzzywuzzy

from fuzzywuzzy import fuzz


similarity_threshold = 70


def group_similar_locations(df, col_name):
    
    groups = {}

    for loc in df[col_name].unique():
        matched = False
        for key in groups.keys():
            
            similarity = fuzz.token_set_ratio(loc, key)
            if similarity > similarity_threshold:
                
                groups[key].append(loc)
                matched = True
                break
        
        
        if not matched:
            groups[loc] = [loc]
    
    
    group_names = {}
    for key in groups.keys():
        for loc in groups[key]:
            group_names[loc] = key
            
    df['group_name'] = df[col_name].apply(lambda x: group_names[x])

    
    return df

```

```python
new_df = group_similar_locations(new_df,"user_location")
```

```python
new_df
```

```python
new_df = new_df.groupby(["group_name"],as_index=False).agg({'values':'sum'})
new_df.sort_values("values",ascending=False, inplace=True)
new_df = new_df.reset_index(drop=True)
new_df
```
